## 3. Visualize Segmented Organs

```python
# store images in environment 
image1 = os.path.join(image_dir, 'LIVER_021_0000.nii.gz')
label1 = os.path.join(label_dir, 'LIVER_021.nii.gz')
image2 = os.path.join(image_dir, 'LIVER_022_0000.nii.gz')
label2 = os.path.join(label_dir, 'LIVER_022.nii.gz')
```

```python
scan1 = sitk.ReadImage(image1)
segmentation1 = sitk.ReadImage(label1)
scan2 = sitk.ReadImage(image2)
segmentation2 = sitk.ReadImage(label2)
```

```python
# Display the images
%matplotlib inline
import matplotlib.pyplot as plt

plt.figure(figsize=(20,20))
# First image
plt.subplot(2,2,1)
plt.imshow(sitk.GetArrayFromImage(scan1)[20,:,:], cmap="gray")
plt.title("Liver #1")
plt.subplot(2,2,2)
plt.imshow(sitk.GetArrayFromImage(segmentation1)[20,:,:])        
plt.title("Segmentation #1")

# Second image
plt.subplot(2,2,3)
plt.imshow(sitk.GetArrayFromImage(scan2)[20,:,:], cmap="gray")        
plt.title("Liver #2")

plt.subplot(2,2,4)
plt.imshow(sitk.GetArrayFromImage(segmentation2)[20,:,:])        
plt.title("Segmentation #2")

plt.show()
```
